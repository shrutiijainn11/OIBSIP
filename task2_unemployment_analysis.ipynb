```python
import datetime as dt
import calendar

df['Date'] = pd.to_datetime(df['Date'], dayfirst = True)
df['month_int'] = df['Date'].dt.month
df['month'] = df['month_int'].apply(lambda x: calendar.month_abbr[x])
```

```python
df_new = df[['State','Estimated_Unemployment_rate']].groupby('State').sum().sort_values('Estimated_Unemployment_rate', ascending = False)
```

```python
fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
df_new[:10].plot(kind='bar', color='brown', figsize=(30,5),ax=ax0)
ax0.set_title('Top 10 states with highest')
ax0.set_xlabel('State')
ax0.set_ylabel('Number of people unemployment %')
```

```python
fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
df1_new[:10].plot(kind='bar', color='pink', figsize=(30,5),ax=ax0)
ax0.set_title('Top 10 states with highest')
ax0.set_xlabel('State')
ax0.set_ylabel('Number of people unemployment %')
```
